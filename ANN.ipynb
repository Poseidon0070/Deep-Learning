```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("./CSVs/churn_modeling.csv")
df.head(5)
```

```python
x = df.iloc[:,3:-1]
x.head(5)
```

```python
y = df.iloc[:,-1]
y.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
geography = pd.get_dummies(x['Geography'])
gender = pd.get_dummies(x['Gender'])
x = pd.concat([x,geography,gender],axis=1)
x.head(5)
```

```python
x = x.drop(['Geography','Gender'],axis=1)
x.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
x_train
```

```python
import keras
```

```python
from tensorflow.keras import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
classifier = keras.Sequential()
```

```python
classifier.add(keras.layers.Dense(6,activation='relu',kernel_initializer='he_uniform',input_shape=(13,)))
```

```python
classifier.add(keras.layers.Dense(6,activation='relu',kernel_initializer='he_uniform'))
```

```python
classifier.add(keras.layers.Dense(1,activation='sigmoid',kernel_initializer='glorot_uniform'))
```

```python
classifier.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model_history = classifier.fit(x_train,y_train,validation_split=0.33,validation_batch_size=10,epochs=100)
```

```python
y_pred = classifier.predict(x_test)
```

```python
for i in range(len(y_pred)):
    if(y_pred[i] >= 0.5):
        y_pred[i] = 1
    else:
        y_pred[i] = 0
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,y_test)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred,y_test)
accuracy
```
